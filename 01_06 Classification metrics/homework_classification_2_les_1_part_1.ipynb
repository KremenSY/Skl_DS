### 1.3
–ó–∞–º–µ–Ω–∏—Ç–µ –≤—Å–µ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∫–æ–ª–æ–Ω–∫–∏ Age –Ω–∞ –º–µ–¥–∏–∞–Ω—ã –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –ø–æ–ª–∞ —á–µ–ª–æ–≤–µ–∫–∞: —Ç.–µ. –µ—Å–ª–∏ –ø–æ–ª —á–µ–ª–æ–≤–µ–∫–∞ –≤ —Å—Ç—Ä–æ–∫–µ —Å –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º "male", –∑–∞–º–µ–Ω—è–µ—Ç–µ –ø—Ä–æ–ø—É—â–µ–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –≤–æ–∑—Ä–∞—Å—Ç–∞ –Ω–∞ –º–µ–¥–∏–∞–Ω—É –ø–æ –≤—Å–µ–º –∏–∑–≤–µ—Å—Ç–Ω—ã–º –≤–æ–∑—Ä–∞—Å—Ç–∞–º –º—É–∂—á–∏–Ω, –∏ –Ω–∞–æ–±–æ—Ä–æ—Ç. –í—ã–≤–µ–¥–∏—Ç–µ –º–µ–¥–∏–∞–Ω—ã –≤–æ–∑—Ä–∞—Å—Ç–∞ –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –ø–æ–ª–∞. –ü—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∫–æ–ª–æ–Ω–æ–∫ Cabin –∏ Embarked –∑–∞–º–µ–Ω–∏—Ç–µ –Ω–∞ U (Unknown).

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ: –¥–ª—è –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∏ –ø–æ –ø–æ–ª—É –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –º–µ—Ç–æ–¥ df.groupby(), –∫–æ—Ç–æ—Ä—ã–π –∏–º–µ–µ—Ç –≤—Å—Ç—Ä–æ–µ–Ω–Ω—ã–µ –∞–≥—Ä–µ–≥–∞—Ç–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏ (–≤ —Ç.—á. –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –º–µ–¥–∏–∞–Ω—ã).  –î–ª—è —É—Å—Ç–∞–Ω–æ–≤–∫–∏ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ –¥–∞—Ç–∞—Å–µ—Ç–µ –≤–æ—Å–ø–æ–ª—å–∑—É–π—Ç–µ—Å—å —Ñ—É–Ω–∫—Ü–∏–µ–π df.apply().

```python
import pandas as pd
import math

df=pd.read_csv("titanic.csv", sep=",")

def fill_NaN_Age(row):
    if row['Sex']=='female' and row['Age']!=row['Age']: # n!=n means n is NaN
        return df.groupby('Sex')['Age'].median()[0]
    elif row['Sex']=='male' and math.isnan(row['Age']):
        return df.groupby('Sex')['Age'].median()[1]
    else:
        return row['Age']
    
df['Age'] = df.apply(lambda row: fill_NaN_Age(row), axis=1)
df['Cabin'] = df.apply(lambda row: 'U' if row['Cabin']!=row['Cabin'] else row['Cabin'] , axis=1)
df['Embarked'] = df.apply(lambda row: 'U' if row['Embarked']!=row['Embarked'] else row['Embarked'] , axis=1)
df.head(20)              
```

### 1.5
–í –∫–æ–ª–æ–Ω–∫–µ Sex –∑–∞–º–µ–Ω–∏—Ç–µ –∑–Ω–∞—á–µ–Ω–∏—è –Ω–∞ 0, –µ—Å–ª–∏ –ø–æ–ª "male", –∏ –Ω–∞ 1, –µ—Å–ª–∏ "female". –í –∫–æ–ª–æ–Ω–∫–µ Embarked –∑–∞–º–µ–Ω–∏—Ç–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã "U", "S", "C", "Q" –Ω–∞ 0, 1, 2, 3 —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ. –û—Ç–±—Ä–æ—Å—å—Ç–µ –∫–æ–ª–æ–Ω–∫–∏ PassengerId, Name, Ticket, Cabin. –í—ã–≤–µ–¥–∏—Ç–µ –ø–µ—Ä–≤—ã–µ 20 —Å—Ç—Ä–æ–∫ –ø–æ–ª—É—á–∏–≤—à–µ–≥–æ—Å—è –Ω–∞–±–æ—Ä–∞ –¥–∞–Ω–Ω—ã—Ö.

```python
df=pd.read_csv("titanic.csv", sep=",")

def fill_NaN_Age(row):
    if row['Sex']=='female' and row['Age']!=row['Age']: # n!=n means n is NaN
        return df.groupby('Sex')['Age'].median()[0]
    elif row['Sex']=='male' and math.isnan(row['Age']):
        return df.groupby('Sex')['Age'].median()[1]
    else:
        return row['Age']
    
df['Age'] = df.apply(lambda row: fill_NaN_Age(row), axis=1)
df['Cabin'] = df.apply(lambda row: 'U' if row['Cabin']!=row['Cabin'] else row['Cabin'] , axis=1)
df['Embarked'] = df.apply(lambda row: 'U' if row['Embarked']!=row['Embarked'] else row['Embarked'] , axis=1)
df.head(20)              


### YOUR CODE HERE ###
df['Sex'] = df.apply(lambda row: 0 if row['Sex']=='male' else 1, axis=1)
def embarked_to_num (row):
    if row['Embarked'].strip()=='U':
        return 0
    elif row['Embarked']=='S':
        return 1
    elif row['Embarked']=='S':
        return 2
    elif row['Embarked']=='S':
        return 3
    else:
        return 5
df['Embarked'] = df.apply(lambda row: embarked_to_num(row), axis=1)
df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
df
df.to_csv('titanic_preprocessed.csv', sep=',', index=False) # write preprocessed data to csv
```

### 1.6
–ü—Ä–∏—Å—Ç—É–ø–∏–º –∫ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—é –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä–æ–≤. **–£—Å–ª–æ–≤–∏–º—Å—è, —á—Ç–æ, –µ—Å–ª–∏ —Ñ—É–Ω–∫—Ü–∏—è –∏–ª–∏ –æ–±—ä–µ–∫—Ç –º–æ–¥–µ–ª–∏ –∏–º–µ—é—Ç –ø–∞—Ä–∞–º–µ—Ç—Ä random_state, —Ç–æ —É—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º –µ–≥–æ —Ä–∞–≤–Ω—ã–º 17 –≤ –∫–∞–∂–¥–æ–º –∏–∑ —Å–ª—É—á–∞–µ–≤.**

–†–∞–∑–¥–µ–ª–∏—Ç–µ –¥–∞–Ω–Ω—ã–µ –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã–π –∏ —Ç–µ—Å—Ç–æ–≤—ã–π –¥–∞—Ç–∞—Å–µ—Ç—ã, —É—Å—Ç–∞–Ω–æ–≤–∏–≤ —Ä–∞–∑–º–µ—Ä —Ç–µ—Å—Ç–æ–≤–æ–≥–æ –∫–∞–∫ 0.25 (–ø–µ—Ä–≤–∞—è –∫–æ–ª–æ–Ω–∫–∞ Survived —è–≤–ª—è–µ—Ç—Å—è —Ü–µ–ª–µ–≤–æ–π, –ø–æ—ç—Ç–æ–º—É –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ —Å–Ω–∞—á–∞–ª–∞ –µ–µ –æ—Ç–¥–µ–ª–∏—Ç—å –æ—Ç –ø—Ä–∏–∑–Ω–∞–∫–æ–≤).

```python
import numpy as np
X=np.array(df.drop(['Survived'], axis=1))
y=np.array(df['Survived'])
```

```python
### YOUR CODE HERE ###
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=17)
```

### 1.7

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
```

```python
# KNN (K Nearest Neighbors)
classifirer_knn = KNeighborsClassifier(n_neighbors=5)
model_knn=classifirer_knn.fit(X_train, y_train)
predictions_knn=classifirer_knn.predict(X_test)
# Gaussian Naive Bayes
classifier_gnb=GaussianNB()
model_gnb=classifier_gnb.fit(X_train, y_train)
predictions_gnb=classifier_gnb.predict(X_test)
# Decision Tree
classifier_dtc=DecisionTreeClassifier(random_state=17)
model_dtc=classifier_dtc.fit(X_train, y_train)
predictions_dtc=classifier_dtc.predict(X_test)
# Logistic regression
classifier_lr=LogisticRegression(solver='lbfgs', random_state=17)
model_lr=classifier_lr.fit(X_train, y_train)
predictions_lr=classifier_lr.predict(X_test)
```

### 1.10
–î–ª—è –∫–∞–∂–¥–æ–≥–æ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä–∞ –≤—ã–≤–µ–¥–∏—Ç–µ –º–∞—Ç—Ä–∏—Ü—É –æ—à–∏–±–æ–∫ –∏ —Å–∞–º–æ—Å—Ç–æ—è—Ç–µ–ª—å–Ω–æ —Ä–∞—Å—Å—á–∏—Ç–∞–π—Ç–µ –º–µ—Ç—Ä–∏–∫—É recall (—Ä–∞—Å—á–µ—Ç—ã –¥–æ–ª–∂–Ω—ã –ø—Ä–∏—Å—É—Ç—Å—Ç–≤–æ–≤–∞—Ç—å). –ü—Ä–æ–≤–µ—Ä—å—Ç–µ —Ä–∞—Å—á–µ—Ç—ã, –≤–æ—Å–ø–æ–ª—å–∑–æ–≤–∞–≤—à–∏—Å—å –≤—Å—Ç—Ä–æ–µ–Ω–Ω–æ–π –º–µ—Ç—Ä–∏–∫–æ–π recall_score.

```python
from sklearn.metrics import confusion_matrix, recall_score
from sklearn.metrics import precision_score
```

```python
def precission_recall (y_test, y_predict):
    TP, FP, FN, TN=0, 0, 0, 0
    for i in range (len(y_test)):
        if y_predict[i]==1 and y_test[i]==1:
            TP+=1
        elif y_predict[i]==1 and y_test[i]==0:
            FP+=1
        elif y_predict[i]==0 and y_test[i]==1:
            FN+=1
        elif y_predict[i]==0 and y_test[i]==0:
            TN+=1
        else: pass
    Precision=TP/(TP+FP)
    Recall=TP/(TP+FN)
    error_matrix=np.array([[TP, FP],[FN, TN]])
    return Precision, Recall, error_matrix

precission_knn, recall_knn, error_matrix_knn=precission_recall(y_test, predictions_knn)
print(f"for kNN classifier precission is {round(precission_knn,3)}, \
recall is {round(recall_knn, 3)}. Error matrix: \n {error_matrix_knn}")
print ("precision (micro) kNN is", round(precision_score(y_test, predictions_knn, average='micro'),3))
print ("precision (macro) kNN is", round(precision_score(y_test, predictions_knn, average='macro'),3))
print ("precision (weighted) kNN is", round(precision_score(y_test, predictions_knn, average='weighted'),3))
print ("recall (micro) kNN is", round(recall_score(y_test, predictions_knn, average='micro'),3))
print ("recall (macro) kNN is", round(recall_score(y_test, predictions_knn, average='macro'),3))
print ("recall (weighted) kNN is", round(recall_score(y_test, predictions_knn, average='weighted'),3))

precission_gnb, recall_gnb, error_matrix_gnb=precission_recall(y_test, predictions_gnb)
print(f"for Gaussian Naive Bayes classifier precission is {round(precission_gnb,3)}, \
recall is {round(recall_gnb, 3)}. Error matrix: \n {error_matrix_gnb} ")
print ("precision (micro) Gaussian Naive Bayes is", round(precision_score(y_test, predictions_gnb, average='micro'),3))
print ("precision (macro) Gaussian Naive Bayes is", round(precision_score(y_test, predictions_gnb, average='macro'),3))
print ("precision (weighted) Gaussian Naive Bayes is", round(precision_score(y_test, predictions_gnb, average='weighted'),3))
print ("recall (micro) Gaussian Naive Bayes is", round(recall_score(y_test, predictions_gnb, average='micro'),3))
print ("recall (macro) Gaussian Naive Bayes is", round(recall_score(y_test, predictions_gnb, average='macro'),3))
print ("recall (weighted) Gaussian Naive Bayes is", round(recall_score(y_test, predictions_gnb, average='weighted'),3))

precission_dtc, recall_dtc,  error_matrix_dtc=precission_recall(y_test, predictions_dtc)
print(f"for Decision Tree classifier precission is {round(precission_dtc,3)}, \
recall is {round(recall_dtc, 3)}. Error matrix: \n {error_matrix_dtc}")
print ("precision (micro) Decision Tree is", round(precision_score(y_test, predictions_gnb, average='micro'),3))
print ("precision (macro) Decision Tree is", round(precision_score(y_test, predictions_gnb, average='macro'),3))
print ("precision (weighted) Decision Tree is", round(precision_score(y_test, predictions_gnb, average='weighted'),3))
print ("recall (micro) Decision tree is", round(recall_score(y_test, predictions_dtc, average='micro'),3))
print ("recall (macro) Decision tree is", round(recall_score(y_test, predictions_dtc, average='macro'),3))
print ("recall (weighted) Decision tree is", round(recall_score(y_test, predictions_dtc, average='weighted'),3))

precission_lr, recall_lr,  error_matrix_lr=precission_recall(y_test, predictions_lr)
print(f"for Logistic regression classifier precission is {round(precission_lr,3)}, \
recall is {round(recall_lr, 3)}. Error matrix: \n {error_matrix_lr}")
print ("precision (micro) Logistic regression is", round(precision_score(y_test, predictions_lr, average='micro'),3))
print ("precision (macro) Logistic regression is", round(precision_score(y_test, predictions_lr, average='macro'),3))
print ("precision (weighted) Logistic regression is", round(precision_score(y_test, predictions_lr, average='weighted'),3))
print ("recall (micro) Logistic regression is", round(recall_score(y_test, predictions_lr, average='micro'),3))
print ("recall (macro) Logistic regression is", round(recall_score(y_test, predictions_lr, average='macro'),3))
print ("recall (weighted) Logistic regression is", round(recall_score(y_test, predictions_lr, average='weighted'),3))
```
